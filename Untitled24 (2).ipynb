```python
import os
import shutil

# Function to organize files by their extensions
def organize_files(directory):
    # Define categories of files and their extensions
    file_categories = {
        'Images': ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff'],
        'Documents': ['.pdf', '.docx', '.txt', '.xlsx', '.pptx'],
        'Audio': ['.mp3', '.wav', '.aac', '.flac'],
        'Videos': ['.mp4', '.avi', '.mov', '.mkv'],
        'Archives': ['.zip', '.tar', '.rar', '.7z'],
        'Code': ['.py', '.java', '.html', '.css', '.js'],
    }

    # Check if the directory exists
    if not os.path.exists(directory):
        print(f"The directory '{directory}' does not exist!")
        return

    # Loop through all the files in the directory
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)

        # If it's a file (not a subdirectory)
        if os.path.isfile(file_path):
            # Get the file extension
            file_extension = os.path.splitext(filename)[1].lower()

            # Check which category the file belongs to
            moved = False
            for category, extensions in file_categories.items():
                if file_extension in extensions:
                    category_path = os.path.join(directory, category)

                    # If the category folder doesn't exist, create it
                    if not os.path.exists(category_path):
                        os.makedirs(category_path)

                    # Move the file to the appropriate folder
                    shutil.move(file_path, os.path.join(category_path, filename))
                    print(f"Moved '{filename}' to '{category}' folder.")
                    moved = True
                    break

            # If the file extension doesn't match any category
            if not moved:
                other_folder = os.path.join(directory, 'Others')
                if not os.path.exists(other_folder):
                    os.makedirs(other_folder)
                shutil.move(file_path, os.path.join(other_folder, filename))
                print(f"Moved '{filename}' to 'Others' folder.")

# Main function to run the program
def main():
    # Get the directory path from the user
    directory = input("Enter the directory path to organize: ")

    # Organize files in the given directory
    organize_files(directory)

# Run the program
if __name__ == "__main__":
    main()
```

```python
# Dictionary to store votes for each candidate
votes = {}

# Function to add a candidate
def add_candidate():
    name = input("Enter the candidate's name: ")
    if name not in votes:
        votes[name] = 0  # Initialize the candidate's vote count to 0
        print(f"Candidate '{name}' added successfully!\n")
    else:
        print(f"Candidate '{name}' already exists.\n")

# Function to cast a vote
def cast_vote():
    print("\nCandidates available for voting:")
    for candidate in votes:
        print(f"- {candidate}")

    # Get the user's vote
    vote = input("\nEnter the name of the candidate you want to vote for: ")

    if vote in votes:
        votes[vote] += 1
        print(f"Vote for '{vote}' has been casted successfully!\n")
    else:
        print(f"Invalid candidate '{vote}', please try again.\n")

# Function to view the current votes
def view_votes():
    if votes:
        print("\nCurrent vote counts:")
        for candidate, vote_count in votes.items():
            print(f"{candidate}: {vote_count} votes")
    else:
        print("No candidates available.\n")

# Function to display the election results
def display_results():
    if votes:
        print("\nElection Results:")
        for candidate, vote_count in votes.items():
            print(f"{candidate}: {vote_count} votes")
        # Find the winner
        winner = max(votes, key=votes.get)
        print(f"\nThe winner is '{winner}' with {votes[winner]} votes!")
    else:
        print("No votes cast yet.\n")

# Main function to run the voting system
def voting_system():
    while True:
        print("\nWelcome to the Voting System!")
        print("1. Add Candidate")
        print("2. Cast Vote")
        print("3. View Votes")
        print("4. Display Results")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            add_candidate()
        elif choice == '2':
            cast_vote()
        elif choice == '3':
            view_votes()
        elif choice == '4':
            display_results()
        elif choice == '5':
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please select a number between 1 and 5.\n")

# Run the voting system
if __name__ == "__main__":
    voting_system()
```
